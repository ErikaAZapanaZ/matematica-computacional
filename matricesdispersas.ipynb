```python
import scipy.sparse as sp
import numpy as np
```

```python
A_dense=np.array([[0,1,0,0],[0,0,0,2],[0,0,3,0],[4,0,0,0]])
print("A:\n", A_dense)
values = [1,2,3,4]
rows = [0,1,2,3]
cols = [1,3,2,0]
A_sparse=sp.coo_matrix((values,(rows,cols)), shape=[4,4])
print("A_sparse:\n", A_sparse)

A_crs = A_sparse.tocsr()
print(A_crs.data)
print(A_crs.indices)
print(A_crs.indptr)
```

```python
data =[1,3,2,5,6]
indices = [2,1,3,2,4]
indptr = [0,2,4,5]
B_csr = sp.csr_array((data,indices,indptr),shape=[3,5])
print("B_csr\n",B_csr)
print("la matriz que se forma:\n",B_csr.toarray())
B_csr.nnz # nùmero de entradas no nulas
```

```python
import matplotlib.pyplot as plt
```

```python
nrows =1000
ncols = 1000
A = sp.random(nrows, ncols,density=0.001, format='csr', data_rvs=np.random.randn)
# A_densa = A.toarray()

C=A.T*A
# print(A.toarray())
# plt.imshow(A_densa, cmap="viridis")
plt.figure(figsize=(5,5))
# plt.spy(A, markersize=1) #  tamaño grande para que se vea bien
plt.spy(C, markersize=1)
#plt.title("Patron de dispersidad")
plt.title(f"Matriz de {nrows}x{ncols} (densidad = {A.nnz/(nrows*ncols):.4f})")
plt.show()
```
